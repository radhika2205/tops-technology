```python
# 1.A bag contains 5 red balls, 4 blue balls, and 3 green balls. One ball is drawn at random. 
# a. What is the probability of getting a red ball? 
# b. What is the probability of getting a red or blue ball? 
# c. What is the probability of not getting a green ball?

Probability is the mathematical measure of how likely an event is to occur, 
red = 5
blue = 4
green = 3
total = red + blue + green

# <!-- (a)-->
p_red = red/total


# (b)
p_red_or_p_blue = (red + blue)/total

# (c)
p_not_green = 1-(green/total)

print(f"probability of red ball = {p_red}")
print(f"probability of red or blue ball = {p_red_or_p_blue}")
print(f"probability of not green ball = {p_not_green}")
```

```python
# 3.You are given the following marks of 20 students: 
# 69,35,26,55,28,99,69,50,31,32,23,28,77,80,75,45,69,61,27,48,97 
# a. Find the mean, median, and mode 
# b. Calculate standard deviation 
# c. Comment on the symmetry of the data distribution.

marks = [69,35,26,55,28,99,69,50,31,32,23,28,77,80,75,45,69,61,27,48,97]

# a.
mean = np.mean(marks)
median = np.median(marks)
mode = stats.mode(marks)

print(f"mean = {mean}")
print(f"median={median}")
print(f"mode={mode}")


# b.
std_dev = np.std(marks)
print(f"standard deviation = {std_dev}")
STD àªàªŸàª²à«‡ àª¶à«àª‚?

Data àª•à«‡àªŸàª²à«àª‚ àª«à«‡àª²àª¾àª¯à«‡àª²à«àª‚ àª›à«‡ (spread) àª àª¬àª¤àª¾àªµà«‡ àª›à«‡.

ðŸ‘‰ Marks àª¬àª§àª¾àª‚ àª¨àªœà«€àª•-àª¨àªœà«€àª• àª›à«‡ àª•à«‡ àª¬àª¹à« àª¦à«‚àª°-àª¦à«‚àª° â€” àª STD àª¬àª¤àª¾àªµà«‡
# c.
skewness = stats.skew(marks)
print(f"skewness = {skewness}")
Skewness àªàªŸàª²à«‡ àª¶à«àª‚?

Data àª•àªˆ àª¬àª¾àªœà« àªà«àª•à«‡àª²à«àª‚ àª›à«‡ àª àª¬àª¤àª¾àªµà«‡.
# comment
if skewness > 0:
    print("the data is right-skewed")
elif skewness < 0:
    print("the data is left-skewed")
else:
    print("the data is symmetric.")
```

```python
# 4. Use the â€œStudent Attitude and Behaviorâ€ dataset and select the following columns: Height, 
# Weight, 10th Marks, 12th Marks. 
# a Find Mean, Median, Mode and Quartiles 
# b Find Skewness, kurtosis and Interpretation Based on the Skewness, kurtosis.

cols = ["Height(CM)", "Weight(KG)", "10th Mark", "12th Mark"]
data = df[cols].apply(pd.to_numeric, errors='coerce').dropna()

# a.
# Mean, Median, Mode
mean_values = data.mean()
median_values = data.median()
mode_values = data.mode().iloc[0]
Mode àªàª•àª¥à«€ àªµàª§àª¾àª°à«‡ àª¹à«‹àªˆ àª¶àª•à«‡
ðŸ‘‰ àªàªŸàª²à«‡ àªªàª¹à«‡àª²à«€ row (first mode) àª²à«‡àªµàª¾ iloc[0]

# Quartiles
Q1 = data.quantile(0.25)
Q2 = data.quantile(0.5)
Q3 = data.quantile(0.75)

print(" Mean:\n", mean_values)
print("\n Median:\n", median_values)
print("\n Mode:\n", mode_values)

print("\n Quartiles:")

print("\n Q1 (25%):\n", Q1)
print("\n Q2 (50%):\n", Q2)
print("\n Q3 (75%):\n", Q3)

# b.
cols = ["Height(CM)", "Weight(KG)", "10th Mark", "12th Mark"]
data = df[cols].apply(pd.to_numeric, errors='coerce').dropna()

# Skewness & Kurtosis

skewness = data.skew()
Data àª•àªˆ àª¬àª¾àªœà« àªà«àª•à«‡àª²à«àª‚ àª›à«‡

> 0 â†’ Right / Positively skewed

< 0 â†’ Left / Negatively skewed

= 0 â†’ Symmetric
kurt = data.kurtosis()
Kurtosis àªàªŸàª²à«‡ àª¶à«àª‚?

ðŸ‘‰ Data àª•à«‡àªŸàª²à«àª‚ sharp àª•à«‡ flat àª›à«‡

Value	Type
> 3	Leptokurtic (sharp peak)
= 3	Mesokurtic (normal)
< 3	Platykurtic (flat)

# data ketlu swed aetle ke zukelu che
# data ketlu flat che - kurtosis
# Leptokurtic
# Mesokurtic
# Platykurtic

print("\n Skewness:\n", skewness)
print("\n Kurtosis:\n", kurt)

print("\n Interpretation:")

for col in data.columns:
    s = skewness[col]
    k = kurt[col]

    # Skewness interpretation
    if s > 0: 
        skew_text = "Positively Skewed"
    elif s < 0:
        skew_text = "Negatively Skewed"
    else:
        skew_text = "Symmetric"

    # Kurtosis interpretation
    if k > 3:
        kurt_text = "Leptokurtic"
    elif k < 3:
        kurt_text = "Platykurtic"
    else:
        kurt_text = "Mesokurtic"

    print(f"\n {col}:")
    print(f"   Skewness = {s:.2f} â†’ {skew_text}")
    print(f"   Kurtosis = {k:.2f} â†’ {kurt_text}")

df[cols] â†’ àª«àª•à«àª¤ àª† 4 columns àª²à«‡ àª›à«‡

pd.to_numeric â†’ àª¬àª§à«àª‚ data number àª®àª¾àª‚ àª«à«‡àª°àªµà«‡ àª›à«‡

errors='coerce' â†’ text / wrong value àª¨à«‡ NaN àª¬àª¨àª¾àªµà«‡

dropna() â†’ NaN àªµàª¾àª³à«€ row àª•àª¾àª¢à«€ àª¨àª¾àª–à«‡
```

```python
# 5. An investment analyst wants to assess the relationship between the stock prices of two 
# companies to identify potential investment opportunities. Data: Let's consider the daily 
# closing prices (in dollars) of Company A and Company B for a sample of 20 trading days:  
# Company A: 45, 47, 48, 50, 52, 53, 55, 56, 58, 60, 62, 64, 65, 67, 69, 70, 72, 74, 76, 77 
# Company B: 52, 54, 55, 57, 59, 60, 61, 62, 64, 66, 67, 69, 71, 73, 74, 76, 78, 80, 82, 83 
# â€¢ Calculate the covariance and correlation between the stock prices of Company A 
# and Company B. Interpret the value of the covariance and explain the nature of the 
# relationship between the two stocks.

company_A = [45, 47, 48, 50, 52, 53, 55, 56, 58, 60, 62, 64, 65, 67, 69, 70, 72, 74, 76, 77]
company_B = [52, 54, 55, 57, 59, 60, 61, 62, 64, 66, 67, 69, 71, 73, 74, 76, 78, 80, 82, 83]

# Create DataFrame
df = pd.DataFrame({'Company A': company_A, 'Company B': company_B})

# Covariance
covariance = df['Company A'].cov(df['Company B'])
àª¬à«‡ variables àª¸àª¾àª¥à«‡-àª¸àª¾àª¥à«‡ àª•à«‡àªµà«€ àª°à«€àª¤à«‡ àª¬àª¦àª²àª¾àª¯ àª›à«‡ àª àª¬àª¤àª¾àªµà«‡

Covariance > 0 â†’ àª¬àª‚àª¨à«‡ àª¸àª¾àª¥à«‡ àªµàª§à«‡ / àª˜àªŸà«‡

Covariance < 0 â†’ àªàª• àªµàª§à«‡, àª¬à«€àªœà«àª‚ àª˜àªŸà«‡

Covariance = 0 â†’ àª•à«‹àªˆ relationship àª¨àª¹à«€àª‚

ðŸ‘‰ àª…àª¹à«€àª‚ prices àª¬àª‚àª¨à«‡ àª¸àª¾àª¥à«‡ àªµàª§à«‡ àª›à«‡,
àªàªŸàª²à«‡ covariance positive àª†àªµàª¶à«‡
# Correlation
correlation = df['Company A'].corr(df['Company B'])
Correlation àªàªŸàª²à«‡ àª¶à«àª‚?

Relationship àª¨à«€ strength àª…àª¨à«‡ direction

Value àª¹àª‚àª®à«‡àª¶àª¾ -1 àª¥à«€ +1 àªµàªšà«àªšà«‡

Value	Meaning
+1	Perfect positive relation
0	No relation
-1	Perfect negative relation

ðŸ‘‰ àª…àª¹à«€àª‚ prices almost same pattern àª®àª¾àª‚ àªµàª§à«‡ àª›à«‡
àªàªŸàª²à«‡ correlation +1 àª¨à«€ àª¨àªœà«€àª• àª†àªµàª¶à«‡

print("Covariance between Company A and B:", covariance)
print("Correlation between Company A and B:", correlation)
```
