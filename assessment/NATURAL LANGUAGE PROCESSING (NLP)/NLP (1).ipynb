```python
import os

# Load dataset from JSON file using the correct path
# The 'path' variable is available from previous executions (cell _aJKAis_Cob8 and XRkc_m_FDGp7)
file_path = os.path.join(path, "News_Category_Dataset_v3.json")

records = []
with open(file_path, "r") as f:
    for line in f:
        records.append(json.loads(line))

df = pd.DataFrame(records)
print("Total Records:", len(df))
print("\nAll Categories:")
print(df["category"].value_counts())
```

1.Keep only these categories: TECHNO, ENTERTAINMENT, POLITICS, BUSINESS.

```python
categories_to_keep = ["TECH", "ENTERTAINMENT", "POLITICS", "BUSINESS"]
df = df[df["category"].isin(categories_to_keep)].reset_index(drop=True)

print("Filtered Records:", len(df))
print("\nFiltered Categories:")
print(df["category"].value_counts())
df.head()
```
