Create a simple Q&A chatbot for Zomato-style restaurant info using any language or platform you know (Python, JavaScript, etc.): given a restaurant name and cuisine, return a canned response like 'Yes, Pizza Palace serves Italian cuisine.'

```python
# Restaurant Database
restaurants = {
    "Pizza Palace": "Italian",
    "Spicy Hub": "Indian",
    "Dragon Bowl": "Chinese",
    "Burger Town": "American"
}

# User Input
restaurant_name = input("Enter restaurant name: ")
cuisine = input("Enter cuisine: ")

# Check Information
if restaurant_name in restaurants:
    if restaurants[restaurant_name].lower() == cuisine.lower():
        print(f"Yes, {restaurant_name} serves {cuisine} cuisine.")
    else:
        print(f"No, {restaurant_name} does not serve {cuisine} cuisine. It serves {restaurants[restaurant_name]} cuisine.")
else:
    print("Restaurant not found in database.")
```

Use GitHub Copilot or ChatGPT to help you write a function that takes a list of song names and returns only those that start with the letter 'S', like a Spotify playlist filter.<br><br><em><strong>Constraint:</strong> You must show the prompt you used and the code Copilot or ChatGPT generated.</em>

Prompt: Write a Python function that takes a list of song names and returns only the songs that start with the letter 'S'. Include an example.

```python
def filter_songs_starting_with_s(song_list):
    return [song for song in song_list if song.startswith('S')]

# Example
songs = [
    "Shape of You",
    "Believer",
    "Senorita",
    "Sunflower",
    "Perfect",
    "Stay"
]

filtered_songs = filter_songs_starting_with_s(songs)

print("Songs starting with 'S':")
print(filtered_songs)
```

1. Incorrect or Hallucinated Responses

Challenge:
Generative AI chatbots may provide incorrect information, such as wrong movie timings on BookMyShow or incorrect product details on Myntra.

Solution:

2. High Response Time and Infrastructure Cost

Challenge:
Large AI models require significant computing resources, which can increase costs and slow down responses when many users access the chatbot simultaneously.

Solution:
Use smaller optimized models, response caching, and cloud auto-scaling to improve speed while reducing operational costs.
