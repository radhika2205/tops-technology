2.
Create a Python class Discriminator with a method discriminate(image) that takes a 28x28 image array and returns a probability score between 0 and 1 indicating if the image is 'real' or 'fake'.

```python
import numpy as np

class Discriminator:

    def discriminate(self, image):

        mean_pixel = np.mean(image)

        # Simple rule:
        if mean_pixel > 0.5:
            return 0.8   # likely real
        else:
            return 0.2   # likely fake

# Example
D = Discriminator()
score = D.discriminate(np.random.rand(28,28))

print("Probability:", score)
```
